```python
def save_responses_to_pdf(filename, responses):
    from reportlab.pdfgen import canvas
    from reportlab.lib.pagesizes import A4
    from reportlab.lib.units import inch
    import textwrap

    doc = canvas.Canvas(filename, pagesize=A4)
    page_width, page_height = A4
    text_x = 50  # Set your desired x-position
    text_object = doc.beginText(text_x, 700)
    section_spacing = 25

    for response in responses:
        for line in response.splitlines():
            # Wrap long lines
            available_width = page_width - text_x * 2
            wrapped_lines = textwrap.wrap(line, width=int(available_width / inch * 10))
            for wrapped_line in wrapped_lines:
                text_object.textLine(wrapped_line)

            if text_object.getY() < 50:
                doc.drawText(text_object)
                doc.showPage()
                text_object = doc.beginText(text_x, 700)

        text_object.textLine("")
        text_object.moveCursor(0, section_spacing)

    doc.drawText(text_object)
    doc.save()



# # Priority one.
# def save_responses_to_pdf(filename, responses):
#     from reportlab.pdfgen import canvas
#     from reportlab.lib.pagesizes import letter
#     from reportlab.lib.units import inch
#     import textwrap

#     doc = canvas.Canvas(filename, pagesize=letter)
#     page_width, page_height = letter
#     text_x = 50  # Set your desired x-position
#     text_object = doc.beginText(text_x, 700)

#     for response in responses:
#         for line in response.splitlines():
#             # Wrap long lines
#             available_width = page_width - text_x * 2  # Account for positioning
#             wrapped_lines = textwrap.wrap(line, width=int(available_width / inch * 10))
#             for wrapped_line in wrapped_lines:
#                 text_object.textLine(wrapped_line)

#             if text_object.getY() < 50:
#                 doc.drawText(text_object)
#                 doc.showPage()
#                 text_object = doc.beginText(text_x, 700)

#     doc.drawText(text_object)
#     doc.save()



# # Almost there. Priority two.

# def save_responses_to_pdf(filename, responses):
#     # ... (imports, etc.) ...

#     doc = canvas.Canvas(filename, pagesize=letter)
#     text_object = doc.beginText(50, 700)  # Start a text object

#     for response in responses:
#         for line in response.splitlines():
#             text_object.textLine(line)  # Add line to the text object
#             if text_object.getY() < 50:  # Check if near bottom of page
#                 doc.drawText(text_object)  # Write the text object
#                 doc.showPage()
#                 text_object = doc.beginText(50, 700)

#     doc.drawText(text_object)  # Write any remaining text
#     doc.save()

save_responses_to_pdf("generated_resume.pdf", [
    response1.text,
    response6.text,
    response2.text,
    response3.text,
    response4.text,
    response5.text
])
```
